```python
stats0 = np.load("stats_v0_1.npy", allow_pickle=True)
stats1 = np.load("stats_v1_1.npy", allow_pickle=True)
stats2 = np.load("stats_v2_1.npy", allow_pickle=True)
```

```python
def plot_stats(idx, title):
    plt.plot(stats0[:,idx].astype(float), label="w/ memory replay")
    plt.plot(stats1[:,idx].astype(float), label="w memory replay")
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# number of stages in episode
plot_stats(0, "Stages in episode")
```

```python
# Cumulative return in episode
plot_stats(1, "Cumulative return in episode")
```

```python
# Discounted return in episode
plot_stats(2, "Discounted return in episode")
```

```python
# Mean return in episode
plot_stats(3, "Mean return in episode")
```

```python
# mean return in lifetime
plot_stats(4, "Mean return in lifetime")
```

```python
# mean return per episode
plot_stats(5, "Mean return per episode")
```
